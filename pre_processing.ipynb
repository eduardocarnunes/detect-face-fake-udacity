```python
#import libraries
import pandas as pd
import numpy as np
import os
from face_recognition.face_recognition_cli import image_files_in_folder
import face_recognition
```

```python
def getFeaturesOpenFace(image_folder):
    """Return
    
    Parameters:
        image_folder (string): Path folder with faces images
    
    Returns:
        dict_features_openface (dictionary): Dictionary where each lane has a float array (128 features)     
    
    """
    
    dict_features_openface = {}
    
    for person in os.listdir(image_folder):   
        if not os.path.isdir(os.path.join(image_folder, person)):            
            continue
        cont = 1
        name = str(person)        
        # Loop through each training image for the current person
        for img_path in image_files_in_folder(os.path.join(image_folder, person)):
            print(img_path)
            image = face_recognition.load_image_file(img_path)  
            #features
            features=face_recognition.face_encodings(image)[0]
            dict_features_openface[name + "_" + str(cont)] = features
            cont = cont +1
            
    return dict_features_openface
```

```python
#get features of real faces 
dict_features_real = getFeaturesOpenFace("dataset-final/face_real")
```

```python
#get features of fake faces paper
dict_features_fake_paper = getFeaturesOpenFace("dataset-final/fake_face_paper")
```

```python
#get features of fake faces digital
dict_features_fake_digital = getFeaturesOpenFace("dataset-final/fake_face_digital")
```

```python
def dict_to_csv(dict_features, nameFile):
    """Return
    
    Parameters:
        dict_features (dictionary): Path folder with faces images
        nameFile (String) : name of .csv
    
    Returns:
        dict_features_openface (dictionary): Dictionary where each lane has a float array (128 features)     
    
    """
    aux = pd.DataFrame.from_dict(dict_features, orient='index')    
    export_csv = aux.to_csv (r'dataset-' + nameFile + '.csv', header=True)
    
    return export_csv
```

```python
#dictonary to csv 
real_faces_aux = dict_to_csv(dict_features_real, "real-faces")
fake_paper_aux = dict_to_csv(dict_features_fake_paper, "fake-faces-paper")
fake_digital_aux = dict_to_csv(dict_features_fake_digital, "fake-faces-digital")
```

```python
#read datasets csv
df_face_real = pd.read_csv('dataset-real-faces.csv', delimiter = ',')
df_fake_paper = pd.read_csv('dataset-fake-faces-paper.csv', delimiter = ',')
df_fake_digital = pd.read_csv('dataset-fake-faces-digital.csv', delimiter = ',')
```
